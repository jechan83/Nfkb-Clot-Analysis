# Load preprocessed data from Chan* and Pan* et al., Nature 2025 (GSE277777).

Processed data from: https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE277777


```python
import scanpy as sc
import numpy as np
import pandas as pd

import scipy
```

```python
import GEOparse
import urllib.request
import os

if not (os.path.exists('./GSE277777_allmice-tumor-IGO15123.h5ad') and os.path.exists('./allmice-tumor-IGO15342.h5ad')):
    gse = GEOparse.get_GEO(geo="GSE277777", destdir="./data")

    # Find and download the files
    if 'supplementary_file' in gse.metadata:
        for file_url in gse.metadata['supplementary_file']:
            if not os.path.exists('./GSE277777_allmice-tumor-IGO15123.h5ad') and file_url.endswith('IGO15123.h5ad'):
                filename = os.path.basename(file_url)
                print(f"Found matching file: {filename}")
                print(f"Downloading from: {file_url}")
                
                # Download the file
                save_path = f"./{filename}"
                urllib.request.urlretrieve(file_url, save_path)
                print(f"Downloaded to: {save_path}")
            elif not os.path.exists('./GSE277777_allmice-tumor-IGO15342.h5ad') and file_url.endswith('IGO15342.h5ad'):
                filename = os.path.basename(file_url)
                print(f"Found matching file: {filename}")
                print(f"Downloading from: {file_url}")
                
                # Download the file
                save_path = f"./{filename}"
                urllib.request.urlretrieve(file_url, save_path)
                print(f"Downloaded to: {save_path}")
        else:
            print("Incorrect files specified")
    else:
        print("No supplementary files found in this record")
```
